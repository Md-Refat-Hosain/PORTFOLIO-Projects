```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# --- IMPORTANT: Replace 'temp_full_dst' with the actual name of your DataFrame ---
# Assuming your DataFrame is named 'df' for simplicity in this code.
# If your DataFrame is 'temp_full_dst', replace 'df' with 'temp_full_dst' below.
# Example: df = temp_full_dst
# For demonstration, creating a dummy DataFrame that matches your column names
# You should load your actual data here.
data = {
    'milage': np.random.randint(1000, 300000, 1000),
    'HP': np.random.randint(50, 1000, 1000),
    'L': np.random.uniform(1.0, 8.0, 1000),
    'Cylinders': np.random.choice([4, 6, 8, 10, 12], 1000, p=[0.5, 0.3, 0.15, 0.03, 0.02]),
    'transmission_speeds': np.random.choice([4, 5, 6, 7, 8, 9, 10], 1000, p=[0.05, 0.1, 0.3, 0.2, 0.2, 0.1, 0.05]),
    'Age': np.random.randint(1, 25, 1000),
    'clean_title_yes': np.random.randint(0, 2, 1000),
    'accident_not_reported': np.random.randint(0, 2, 1000),
    'brand_encoded': np.random.rand(1000),
    'model_encoded': np.random.rand(1000),
    'ext_col_encoded': np.random.rand(1000),
    'int_col_encoded': np.random.rand(1000),
    'CVT': np.random.randint(0, 2, 1000),
    'Manual': np.random.randint(0, 2, 1000),
    'other': np.random.randint(0, 2, 1000),
    'Electric': np.random.randint(0, 2, 1000),
    'Flex': np.random.randint(0, 2, 1000),
    'Gasoline': np.random.randint(0, 2, 1000),
    'Hybrid': np.random.randint(0, 2, 1000),
    'Plug-In Hybrid': np.random.randint(0, 2, 1000),
    'â€“': np.random.randint(0, 2, 1000),
    'price': np.random.uniform(5000, 100000, 1000) # Dummy price for completeness
}
df = temp_full_dst


# List of numerical features to check for outliers
# numerical_features = ['milage', 'HP', 'L', 'Cylinders', 'transmission_speeds', 'Age']

print("--- Outlier Detection for Numerical Features ---")

for feature in columns_to_scale:
    print(f"\nAnalyzing feature: '{feature}'")

    # 1. Visual Inspection: Box Plot
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df[feature])
    plt.title(f'Box Plot of {feature}')
    plt.ylabel(feature)
    plt.show()

    # 2. Quantitative Identification: IQR Method
    Q1 = df[feature].quantile(0.25)
    Q3 = df[feature].quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Identify outliers
    outliers_high = df[df[feature] > upper_bound]
    outliers_low = df[df[feature] < lower_bound]

    print(f"  Q1 (25th percentile): {Q1:.2f}")
    print(f"  Q3 (75th percentile): {Q3:.2f}")
    print(f"  IQR: {IQR:.2f}")
    print(f"  Upper Bound for outliers: {upper_bound:.2f}")
    print(f"  Lower Bound for outliers: {lower_bound:.2f}")

    print(f"  Number of high outliers ({feature} > {upper_bound:.2f}): {len(outliers_high)}")
    if not outliers_high.empty:
        print(f"    Sample of high outlier values:\n{outliers_high[feature].head()}")

    print(f"  Number of low outliers ({feature} < {lower_bound:.2f}): {len(outliers_low)}")
    if not outliers_low.empty:
        print(f"    Sample of low outlier values:\n{outliers_low[feature].head()}")

    print("-" * 50)

print("\n--- Reminder: Price (target variable) is log-transformed. ---")
print("Outliers in the log-transformed price are generally considered genuine extreme values")
print("and typically should NOT be removed or capped, as they represent the full range of car prices.")
```
